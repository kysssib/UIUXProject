```python
import requests

def now_location(API_KEY):
  # ipstack API 호출
  response = requests.get(f"http://api.ipstack.com/check?access_key={API_KEY}")
  result = response.json()

  return result['latitude'], result['longitude']

def google_station(API_KEY, latitude, longitude, radius):
  # 가까운 지하철 역 검색
  response = requests.get(f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius={radius}&type=subway_station&key={API_KEY}")
  result = response.json()

  if result['status'] == 'ZERO_RESULTS':
      print('가까운 지하철역을 찾을 수 없습니다.')
  else:
      subway_name = result['results'][0]['name']
      print(subway_name)

def kakao_station(API_KEY, latitude, longitude, radius):
  url = f"https://dapi.kakao.com/v2/local/search/category.json?category_group_code=SW8&x={longitude}&y={latitude}&radius={radius}"
  headers = {"Authorization": f"KakaoAK {API_KEY}"}
  # 가까운 지하철 역 검색
  response = requests.get(url, headers=headers)
  result = response.json()

  if not result['documents']:
      print('가까운 지하철역을 찾을 수 없습니다.')
  else:
      subway_name = result['documents'][0]['place_name']
      print(subway_name)
```
